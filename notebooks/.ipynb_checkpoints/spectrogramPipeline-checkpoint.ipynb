```python
# FFT and Spectrogram creation
fs, test_data = wav.read(wav_files[0])

shape = [64,512] # 64 frequency bins ('rows'); 512 time frames ('columns')
win_len = 512
n_fft = win_len # size of the FFT window we will pass over
hop_len = 384 # hard coding to ensure 25% overlap
overlap = win_len - hop_len # always 128

freqs, times, specs = spectrogram(
    x = test_data,
    fs = fs,
    window = windows.hann(win_len),
    nperseg=win_len,
    noverlap=overlap,
    nfft=n_fft)

plt.figure()
plt.pcolormesh(times, freqs, 10*np.log10(specs), shading='gouraud') # I expect the 10*log10() is actually preprocessing step we will add later.
# Without the log scaling, we will see a very faint image, because each amplitude by itself does not contribute greatly to the overall signal
plt.title("Pipeline Spectrogram")
plt.xlabel("Time [s]")
plt.ylabel("Frequency [Hz]")
```

```python

def scaling (spec, rate, hop_length, gain=0.8, bias=10, power=0.25, t=0.060, eps=1e-6, scale='log'):
    if scale == 'log':
        mel_scaled = librosa.power_to_db(spec**2, ref=np.max, top_db=100)
    elif scale == 'pcen':
        # This is taken directly from the BirdNET codebase
        # It is their means of per-channel energy normalisation
        s = 1 - np.exp(- float(hop_length) / (t * rate))
        M = scipy.signal.lfilter([s], [1, s - 1], spec)
        smooth = (eps + M)**(-gain)
        mel_scaled = (spec * smooth + bias)**power - bias**power
    return mel_scaled
```

```python
def clip_and_normalise (spec,shape=[64,512]):
    # Trim to desired size if too large
    nc_spec = spec[:shape[0],:shape[1]]
    # Normalise to between 0 and 1
    nc_spec -= nc_spec.min()
    if not nc_spec.max() == 0:
        nc_spec /= nc_spec.max()
    else:
        nc_spec = np.clip(nc_spec, 0, 1)
    return nc_spec
    
```
